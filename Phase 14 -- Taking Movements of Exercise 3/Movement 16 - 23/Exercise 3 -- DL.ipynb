# Initializing Inputs and Outputs Variables

```python
X = raw_data.iloc[:,0:10]
y = raw_data.stimulus
```

# One Hot Encoding

```python
num_classes = 8
```

```python
# One hot encoding for categorical labels
import keras
from keras import utils as np_utils

#y = keras.utils.to_categorical(y, num_classes)
y = keras.utils.np_utils.to_categorical(y, num_classes)
```

```python
input_dim = X.shape[1]
```

# Standard Scaler

```python
from sklearn.preprocessing import StandardScaler
standardscaler=StandardScaler()
```

# Splitting data for training and testing

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, random_state = 39)
```

# Deep Learning Architecture

```python
from tensorflow.keras.utils import plot_model
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input
from tensorflow.keras.layers import Dense
from keras.models import Sequential
from keras.layers import Dense, BatchNormalization, Dropout
from tensorflow.keras.optimizers import Adam, SGD, RMSprop, Adadelta, Adagrad, Adamax, Nadam, Ftrl
```

# 1. Define Network

```python
num_classes = 8
input_dim = X.shape[1]
```

```python
visible = Input(shape=(input_dim,))
hidden1 = Dense(3000, activation='relu')(visible)
hidden2 = Dense(1500, activation='relu')(hidden1)
hidden3 = Dropout(0.2)(hidden2)
hidden4 = Dense(750, activation='relu')(hidden3)
hidden5 = Dense(375, activation='relu')(hidden4)
hidden6 = Dense(48, activation='relu')(hidden5)
output = Dense(num_classes, activation='softmax')(hidden6)
model = Model(inputs=visible, outputs=output)
```

```python
model.summary()
```

# 2. Compile Network

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
plot_model(model, 'EMG_ANN.png', show_shapes = True)
```

# 3. Fit Network

```python
import time
start_time = time.time()

history = model.fit(X_train, y_train, batch_size=9000, epochs=300, validation_data = (X_test, y_test))

print (time.time() - start_time, "seconds")
```

# 4.Evaluate Network

# 4.1. Plotting Loss and AccuracyÂ¶

```python
#Loss Curve


plt.plot(figsize = [8,6])

plt.plot(history.history['loss'],'r',linewidth = 3.0)
plt.plot(history.history['val_loss'],'b',linewidth = 3.0)

plt.legend(['Training Loss', 'Validation Loss'], fontsize = 18)

plt.xlabel('Epochs', fontsize = 16)
plt.ylabel('Loss', fontsize = 16)

plt.title('Loss Curve of Deep Neural Netowork' , fontsize = 16 )

```

```python
#Accuracy Curve

plt.plot(figsize = [8,6])
plt.plot(history.history['accuracy'],'r',linewidth = 3.0)
plt.plot(history.history['val_accuracy'],'b',linewidth = 3.0)

plt.legend(['Training Accuracy', 'Validation Accuracy'], fontsize = 18)

plt.xlabel('Epochs', fontsize = 16)
plt.ylabel('Accuracy', fontsize = 16)


plt.title('Accuracy Curve of Deep Neural Netowork' , fontsize = 16 )
```
