```python
X = raw_data.iloc[:,0:10]
y = raw_data.stimulus
```

```python
# Splitting data for training and testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 39)
```

# Importing Libraries of Deeplearning

```python
import tensorflow as tf
import keras
from keras.layers import Dense
from keras.models import Sequential
from keras.layers import Dropout

from keras.callbacks import LearningRateScheduler

import numpy as np
from matplotlib import pyplot as plt

from sklearn.model_selection import GridSearchCV
from keras.constraints import maxnorm
from keras.optimizers import SGD
```

```python
num_classes = 13
```

```python
# One hot encoding for categorical labels
y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)
```

```python
#Take a subset of train for grid search. Let us take 10% for now
from sklearn.model_selection import train_test_split
x_grid, x_not_use, y_grid, y_not_use = train_test_split(X_train, y_train, test_size=0.9, random_state=42)
```

```python
# build the model
input_dim = x_grid.shape[1]
```

```python
model = Sequential()
model.add(Dense(64, activation='relu', kernel_initializer='he_uniform',input_dim = input_dim)) 
model.add(Dense(64, kernel_initializer='he_uniform', activation='relu'))
model.add(Dense(num_classes, kernel_initializer='he_uniform', activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(optimizer = SGD(lr=0.1, momentum=0.3), loss = 'categorical_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(x_grid, y_grid, batch_size=64, epochs=150)
```

```python
test_loss = model.evaluate(x_not_use, y_not_use)
```
